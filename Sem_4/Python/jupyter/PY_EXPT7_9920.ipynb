```python
#q1
class FoodDeliverySystem:
    def __init__(self):
        self.orders_queue = []
        self.prepared_orders = []
        self.delivered_orders = []

    def place_order(self, order):
        self.orders_queue.append(order)
        print(f"Order {order} placed successfully.")

    def prepare_order(self):
        if self.orders_queue:
            prepared_order = self.orders_queue.pop(0)
            self.prepared_orders.append(prepared_order)
            print(f"Order {prepared_order} prepared and ready for delivery.")
        else:
            print("No orders to prepare.")

    def deliver_order(self):
        if self.prepared_orders:
            delivered_order = self.prepared_orders.pop(0)
            self.delivered_orders.append(delivered_order)
            print(f"Order {delivered_order} delivered successfully.")
        else:
            print("No orders to deliver.")

    def display_order_status(self):
        print("Orders Awaiting Preparation:")
        print(self.orders_queue)
        print("Orders Being Prepared:")
        print(self.prepared_orders)
        print("Orders Ready for Delivery:")
        print(self.delivered_orders)


def main():
    system = FoodDeliverySystem()

    while True:
        print("\nFood Delivery Management System")
        print("1. Place Order")
        print("2. Prepare Order")
        print("3. Deliver Order")
        print("4. Display Order Status")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            order = input("Enter the order details: ")
            system.place_order(order)
        elif choice == '2':
            system.prepare_order()
        elif choice == '3':
            system.deliver_order()
        elif choice == '4':
            system.display_order_status()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```

```python
#q2
class Node:
    def __init__(self, item_name, quantity, price):
        self.item_name = item_name
        self.quantity = quantity
        self.price = price
        self.next = None

class ShoppingCart:
    def __init__(self):
        self.head = None

    def add_item(self, item_name, quantity, price):
        new_node = Node(item_name, quantity, price)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def remove_item(self, item_name):
        current = self.head
        if current is None:
            print("The cart is empty.")
            return
        if current.item_name == item_name:
            self.head = current.next
            return
        prev = None
        while current and current.item_name != item_name:
            prev = current
            current = current.next
        if current is None:
            print("Item not found in the cart.")
            return
        prev.next = current.next

    def display_items(self):
        current = self.head
        if current is None:
            print("The cart is empty.")
            return
        print("Items in the cart:")
        while current:
            print(f"Item: {current.item_name}, Quantity: {current.quantity}, Price: {current.price}")
            current = current.next

    def calculate_total(self):
        total = 0
        current = self.head
        while current:
            total += current.quantity * current.price
            current = current.next
        return total

# Example usage:
cart = ShoppingCart()
cart.add_item("Apple", 2, 0.5)
cart.add_item("Banana", 3, 0.3)
cart.add_item("Orange", 1, 0.4)
cart.display_items()
print("Total amount:", cart.calculate_total())

cart.remove_item("Banana")
cart.display_items()
print("Total amount:", cart.calculate_total())
```

```python
#Postlab-1
# What is Doubly link list. Write a python program to create a doubly linked list and add a new node after a given node , traverse a doubly linked list.
# ANS: A doubly linked list is a type of linked list in which each node contains a data part and two pointers.
# The first pointer points to the next node in the sequence, and the second points to the previous node.
# This allows for traversal of the list in both forward and backward directions, which is a significant advantage over singly linked lists.

#Code:
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Function to add a node at the end of the list
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # Function to add a node after a specific node
    def add_after_node(self, key, data):
        cur = self.head
        while cur:
            if cur.data == key and cur.next:
                new_node = Node(data)
                next_node = cur.next
                cur.next = new_node
                new_node.next = next_node
                new_node.prev = cur
                next_node.prev = new_node
                return
            elif cur.data == key and cur.next is None:  # Add at the end
                self.append(data)
                return
            cur = cur.next

    # Function to traverse the doubly linked list
    def print_list(self):
        cur = self.head
        while cur:
            print(cur.data, end=" ")
            cur = cur.next
        print()

# Example usage
dll = DoublyLinkedList()
dll.append(10)
dll.append(20)
dll.append(30)

print("Original list:")
dll.print_list()

# Adding a new node after node with data 20
dll.add_after_node(20, 25)
print("List after adding 25 after 20:")
dll.print_list()

```

```python
#Postlab-2
#Write a python program to implement linked list using collection.dequeue().
from collections import deque

class LinkedList:
    def __init__(self):
        self.linked_list = deque()

    def append(self, data):
        """Append data to the end of the list."""
        self.linked_list.append(data)

    def append_left(self, data):
        """Append data to the beginning of the list."""
        self.linked_list.appendleft(data)

    def insert(self, position, data):
        """Insert data at a specific position.
        If the position is beyond the bounds of the current list, the data will be appended."""
        if position < 0 or position >= len(self.linked_list):
            print("Position out of bounds, appending data at the end.")
            self.append(data)
        else:
            self.linked_list.insert(position, data)

    def delete(self, data):
        """Delete the first occurrence of data in the list."""
        try:
            self.linked_list.remove(data)
        except ValueError:
            print(f"Value {data} not found in the list.")

    def display(self):
        """Display the contents of the list."""
        print("Linked List Contents:", list(self.linked_list))

# Example usage
ll = LinkedList()

# Appending elements
ll.append(10)
ll.append(20)
ll.append(30)
print("After appending 10, 20, 30:")
ll.display()

# Adding an element to the beginning
ll.append_left(5)
print("After adding 5 at the beginning:")
ll.display()

# Inserting an element at a specific position
ll.insert(2, 15)
print("After inserting 15 at position 2:")
ll.display()

# Deleting an element
ll.delete(20)
print("After deleting 20:")
ll.display()
```
